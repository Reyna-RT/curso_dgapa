# Oscilador armónico en 1 dimensión.

Básicamente trata de una masa, $m$, anclada a un resorte con fuerza restitutiva proporcional a su posición. Entoces la fuerza se puede modelar como:

$$ F = -kx $$

El objetivo es averiguar la solución del sistema

```python
#librerias a utilizar
from sympy import *
from sympy.plotting import *
init_printing()
%matplotlib inline
```

```python
m,k,t = symbols('m k t')
x = Function('x')
```

Entonces la ecuación diferencial a resolver es:
$$\frac{d^2 x}{dt^2} + \frac{k}{m}x=0$$
vamos a asumir que $\omega = \sqrt{\frac{k}{m}}$ y es positiva

```python
w = Symbol('\omega',positive=True)
ec_newton=Eq(x(t).diff(t,t)+(w**2)*x(t),0)
solucion=dsolve( ec_newton , x(t))
solucion
```

Si $x(0)=0$ y $\frac{dx(0)}{dt}=1$ entonces las constantes $C_1$ y $C_2$ se determinan de la siguiente manera

```python
C1,C2 = symbols('C_1 C_2')
solucion = C1*sin(w*t)+C2*cos(w*t)
```

```python
solucion.subs(t,0)
```

```python
solucion.diff(t).subs(t,0)
```

Entonces $C_2 = 0$ y $C_1 = \frac{1}{\omega}$

```python
C2=0
C1=1/w
```

```python
sol_final=C1*sin(w*t)+C2*cos(w*t)
sol_final
```

Supongamos que $\omega =2$

```python
caso1 = sol_final.subs(w,2)
caso1
```

```python
g1 = plot(caso1,(t,0,5*pi),show=False,line_color='r')
g2 = plot(caso1.diff(t),(t,0,5*pi),show=False,line_color='g')
g1.extend(g2)
g1.show()
```
