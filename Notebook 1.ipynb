# Caminante aleatorio

Este consiste en lanzar una moneda a cada paso y tomar una decisi√≥n hacia donde caminar. En este notebook se presenta una manera muy sencilla de hacerlo.

# Caminante aleatorio

Prueba de cambios en Git ajeno

```python
from pylab import *
%matplotlib inline
```

```python
trayectoria = zeros(3000)
posicion = 0
for n in range(len(trayectoria)):
    trayectoria[n]=posicion
    volado = rand()
    if volado < 0.5:
        posicion = posicion - 1
    else:
        posicion = posicion + 1
plot(range(len(trayectoria)),trayectoria)
show()
```
